지도학습 : 입력과 출력 샘플 데이터가 있고, 주어진 입력으로부터 출력을 예측하고자 할 때 사용한다.
지도학습은 분류와 회귀로 구성된다.

분류: 미리 정의된 가능성 있는 여러 클래스 레이블 중 하나를 예측

회귀: 연속적인 숫자, 또는 부동 소수점을 예측하는 것

과적합 : 학습할 때 사용한 데이터에 너무 최적화되어 있어서 이후에 들어오는 데이터에 대해 예측을 잘못하는 경우

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### forge 데이터
임의로 발생시킨 데이터

분류 모델 테스트

```python
df1 = pd.read_csv('forge.csv')
df1.head()
```

```python
sub_df1 = df1[df1['target'] == 0]
sub_df2 = df1[df1['target'] == 1]

plt.scatter(sub_df1['attr1'], sub_df1['attr2'], label='0')
plt.scatter(sub_df2['attr1'], sub_df2['attr2'], label='1')
plt.legend()
plt.show()
```

### wave

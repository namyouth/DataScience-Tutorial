페이지 소스보기를 했을 때 우리가 원하는 데이터가 존재한다.

```python
import bs4
import requests
import time
import pandas
```

페이지 요청하는 함수

```python
def getHTML(site):
    # 요청한다.
    response = requests.get(site)
    # 응답결과 html을 반환한다.
    return response.text
```

주어진 html 태그를 분석해서 원하는 데이터를 반환한다.

```python
def getData(html):
    # 수집한 데이터를 담을 리스트
    data_list = []
    
    while True:
        # 딜레이 1초
        time.sleep(1)
        # 요청한다.
        target = getHTML(html)
        
        # bs4 객체를 만든다.
        soup = bs4.BeautifulSoup(target,'lxml')
        # 문서내의 div 태그를 가져온다.
        div_list = soup.find_all('div', class_='p')
        for div_tag in div_list:
            # print(div_tag)
            # print('-'*30)
            # 첫번째 문자열을 추출합시다.
            h3_tag = div_tag.find('h3')
            a_tag = h3_tag
            subject1 = a_tag.text
            # print(subject1)
            # print('-'*30)

            # 두번째 문자열을 추출한다.
            h4_tag = div_tag.find('h4')
            subject2 = h4_tag.text
            # print(subject2)
            # print('-'*30)

            # 날짜와 작가를 추출한다.
            p_tag = div_tag.find('p', class_='author')
            span_tag = p_tag.find('span')
            str1 = span_tag.text
            str2 = str1.split('|')

            str_date = str2[0]
            author = str2[1]
            #print(str_date, author)
            # print('-'*30)

            tuple1 = subject1, subject2, str_date, author
            data_list.append(tuple1)

        # 다음 페이지 처리
        # ul 태그를 가져온다.
        ul_tag = soup.find('ul', class_='pager')
        next_tag = ul_tag.find(any, text='Next')
        # print(next_tag)
        # 태그 이름으로 분기한다.
        if next_tag.name == 'a':
            # 다음 페이지의 주소를 가져온다.
            next_url = next_tag.attrs['href']
            html = f'https://pjt3591oo.github.io{next_url}'
            print(html)
            print(target)
        else:
            break
            print(next_url)
                                 
    
    
    return data_list
        
```

```python
site = ' https://pjt3591oo.github.io/'
data_list = getData(site)
data_list
```

```python
# 저장
columns = ('subject1', 'subject2', 'date', 'author')
df1 = pandas.DataFrame(data_list, columns=columns)
df1.to_csv('ex1.csv', index ='False', encoding='utf-8-sig')
```

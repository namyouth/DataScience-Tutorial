```python
# 기본 폰트 설정 때문에...
import matplotlib as mpl
import matplotlib.font_manager as fm
from matplotlib import rcParams

import matplotlib.pyplot as plt
import numpy as np
```

```python
# 기본 폰트 설정
plt.rc('font', family='Malgun Gothic')
plt.rc('font', size='16')
mpl.rcParams['axes.unicode_minus'] = False
```

### 바 차트
막대 그래프

bar - 수직 그래프

bart - 수평 그래프

```python
x = [1,2,3]
y = [2,3,1]

plt.title('기본 bar 차트')
# plt.bar(x,y)
plt.bar(1, 2)
plt.bar(2, 3)
plt.bar(3, 1)

# tick 설정
plt.xticks(x, ['X1','X2','X3'])
plt.yticks(sorted(y), ['one', 'two', 'three'])

# 각 층의 레이블
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

```python
x = [1, 2, 3]
y = [2, 3, 1]

plt.title('기본 bar 차트')
# plt.barh(x, y)
plt.barh(1, 2)
plt.barh(2, 3)
plt.barh(3, 1)

# tick 설정
plt.yticks(x, ['X1', 'X2', 'X3'])
plt.xticks(sorted(y), ['one', 'two', 'three'])

# 각 축의 레이블
plt.ylabel('x축')
plt.xlabel('y축')

plt.show()
```

### 선 그래프

```python
x = [10,20,30,40]
y = [1,4,9,16]

plt.title('Plot')
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
y2 = np.sin(x)

plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

### 스템플롯

```python
x= -np.linspace(-np.pi, np.pi, 20)
y=np.cos(x)
plt.stem(x,y, '--', use_line_collection=True)
plt.show()
```

### pi 플롯
카테고리

```python
data1 = [15, 30, 45, 10]
plt.title('pie plot')

# 튀어 나오는것 설정
option1 = [0, 0.1, 0, 0]
#각 조각의 레이블
option2 = ['조각1', '조각2', '조각3', '조각4' ]
# 각 조각별 색상
option3 = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# explode : 튀어 나오는 정도
# labels : 각 조각의 label
# colors : 색상
# autopct : 표시할 퍼센트 양식
plt.pie(data1, explode=option1, labels=option2, colors=option3, autopct='%1.1f%%',
        shadow=True, startangle=90, textprops={'size': 10}, counterclock=False)
plt.show()
```

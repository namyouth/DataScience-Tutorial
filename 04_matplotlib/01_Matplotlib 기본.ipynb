```python
import numpy as np
import matplotlib.pyplot as plt
```

### 기본 plot 그리기

```python
plt.title('Plot')
plt.plot([1,4,9,16])
plt.show()
```

```python
array1 = np.array([1,4,9,16])
plt.title('Plot')
plt.plot(array1)
plt.show()
```

### x축, y축 데이터 셋팅

```python
list1 = [10, 20, 30, 40]
list2 = [1, 4, 9, 16]

plt.title('Plot')
# 첫 번째 리스트가 x 축, 두 번째 리스트가 y 축에 셋팅된다.
plt.plot(list1, list2)
plt.show()
```

```python
list1 = [1,1,1,1]
list2 = [2,4,6,8]
list3 = [3,6,9,12]
list4 = [4,8,12,16]

list5 = [1,2,3,4]

plt.title('plot')
# 첫번재 리스트가 x축, 두번재 리스트가 y축에 셋팅된다.
# x축의 리스트들의 첫번째 데이터를 모아 선 하나
# 두번재 데이터들을 모아서 선 하나
plt.plot([list1, list2, list3, list4], list5)
plt.show()
```

```python
list1 = [1,1,1,1]
list2 = [2,4,6,8]
list3 = [3,6,9,12]
list4 = [4,8,12,16]

list5 = [1,2,3,4]

plt.title('plot')
# 첫 번째 리스트가 x 축, 두 번째 리스트가 y 축에 셋팅된다.
# plot의 수 만큼 그래프를 그린다.
plt.plot(list1, list5)
plt.plot(list2, list5)
plt.plot(list3, list5)
plt.plot(list4, list5)

plt.show()
```

```python
array1 = np.array([[1,1,1,1],
                  [2,4,6,8],
                  [3,6,9,12],
                  [4,8,12,16]]) # x축

array2 = np.array([1,2,3,4])   # y축

plt.title('plot')
# 첫번재 리스트가 x축, 두번재 리스트가 y축에 셋팅된다.
# x축의 리스트들의 첫번째 데이터를 모아 선 하나
# 두번재 데이터들을 모아서 선 하나
plt.plot(array1, array2)

plt.show()
```

```python
array1 = np.array([[1,1,1,1],
                  [2,4,6,8],
                  [3,6,9,12],
                  [4,8,12,16]]) # x축

array2 = np.array([1,2,3,4])   # y축

plt.title('plot')
# 첫번재 리스트가 x축, 두번재 리스트가 y축에 셋팅된다.

plt.plot(array1[0], array2)
plt.plot(array1[1], array2)
plt.plot(array1[2], array2)
plt.plot(array1[3], array2)

plt.show()
# 내가 지금 작업하고 있는 것에 따라 그리면 된다. 굳이 numpy->list나 list->numpy로 변경 필요X 
```

### 기본 폰트 설정

```python
from matplotlib import rcParams
plt.rc('font', family='Malgun Gothic')    # 윈도우
# plt.rc('font', family='AppleGothic')      # 맥
plt.rc('font', size='16')

# 사용하는 폰트에 - 가 없을 경우 - 는 지정된 폰트가 아닌
# 원래의 기본 폰트를 사용하도록 지정한다.
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
```

```python
plt.title('한글 타이틀')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel('X축 레이블')
plt.ylabel('Y축 레이블')
plt.show()
```

### 폰트 객체 직접 만들기
사용하고자 하는 폰트가 기본 폰트로 설정되지 않거나 특정 부분에서만
사용하고 싶을 때.

```python
font_path = 'C:/Windows/Fonts/HANDotumExt.ttf'
font_obj = fm.FontProperties(fname=font_path, size=16)
```

```python
plt.title('한글 타이틀')
plt.plot([10,20,30,40], [1,4,9,16])
plt.xlabel('X축 레이블', FontProperties=font_obj)
plt.ylabel('Y축 레이블', FontProperties=font_obj)
plt.show()
```

```python
plt.title('선 스타일(색상, 마커종류, 선종류)')
plt.plot([10,20,30,40],[1,4,9,16], 'r*--')
plt.plot([10,20,30,40],[16,9,4,1], 'bs:')
plt.show()
```

### 상세설정
<img src='image/4.png'/>

```python
plt.title('상세설정')
# c: 선색상, lw: 선굵기, ls: 선종류, marker: 마커종류
# ms: 마커의 크기, mec: 마커 외곽선의 색상,
# mew: 마커 외곽선의 굵기, mfc: 마커 내부의 색상
plt.plot([10,20,30,40], [1,4,9,16], c='b',lw=5,ls='--',
        marker='d', ms=15, mec='g', mew=5, mfc='r')
plt.show()
```

### 범위지정

```python
plt.title('범위지정')
plt.plot([10,20,30,40], [1,4,9,16], c='b',lw=5,ls='--',
        marker='d', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(0, 50)
plt.ylim(-10,30)
plt.show()
```

### 틱 설정

```python
plt.title('틱설정')
plt.plot([10,20,30,40], [1,4,9,16])
plt.xticks([10,20,30,40],['십','이십','삼십','사십'])
plt.yticks([1,4,9,16],['일','사','구','십육'])

plt.show()
```

### 그리드

```python
plt.title('그리드')
plt.plot([10,20,30,40], [1,4,9,16])
plt.grid(True)
plt.show()
```

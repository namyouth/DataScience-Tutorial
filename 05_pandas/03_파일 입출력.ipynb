### 파일에서 데이터를 읽어와 DataFrame 을 생성한다.

```python
# 파일명만 입력하면 첫 번째 줄의 데이터를 컬럼의 이름으로 사용한다.
df1 = pd.read_csv('sample.csv')
df1
```

```python
# 컬럼의 이름을 정해주면 첫번째 줄부터 데이터로 인식한다.
df2 = pd.read_csv('sample.csv', names=['a1','a2','a3'])
df2
```
